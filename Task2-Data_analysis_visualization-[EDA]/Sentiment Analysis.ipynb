```python
#import libraries

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import re
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    score = analyzer.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
    print(score['neg'])
    print(score['pos'])
    print(score['neu'])
    return score['compound']
```

```python
## add columns year and Month

df['year'] = pd.to_datetime(df['date']).dt.year
df['month'] = pd.to_datetime(df['date']).dt.month
df
```

```python
df3 = pd.DataFrame(columns=['year','month','text'])
```

```python
df3['year'] = df.year
df3['month'] = df.month
```

```python
df3['text']=df['parse_tweet'].apply(lambda x:(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",x).split())))
```

```python
df3
```

```python
df3['sentiment']=df3['text'].apply(lambda x:sentiment_analyzer_scores(x))
```

```python
df3
```

```python
df3['sentiment_comp']=df3['sentiment'].apply(lambda c: 'pos' if c >=0 else 'neg')
df3['sentiment_comp'].value_counts()
```

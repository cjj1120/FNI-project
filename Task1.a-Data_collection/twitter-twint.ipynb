```python
# Looping through Keywords list, combine all into a file 
def tweet_twint(date, date2):
    c = twint.Config()
    c.Search = "inflation"
    c.Since = date
    c.Until= date2   # "2021-01-31"
    c.Lang="en"
    c.Filter_retweets = True 
    c.Hide_output = True
    c.Count = True
    c.Pandas = True
    c.Stats = True
    c.Limit = 5800
    twint.run.Search(c)
    df = twint.storage.panda.Tweets_df
    return df        
    
```

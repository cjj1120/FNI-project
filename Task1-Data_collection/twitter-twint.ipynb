```python
keyword_list = ['inflation']
```

```python
# Looping through Keywords list, combine all into a file 
def tweet_twint(keyword_list):
    for word in keyword_list:
        c = twint.Config()
        c.Search = word #search keyword
        c.Since = '2021-01-01'
        c.Until="2021-12-31"
        c.Geo="1.290270,103.851959,20km"
        c.Lang="en"
        c.Filter_retweets = True 
        c.Hide_output = True
        c.Count = True
        c.Pandas = True
        c.Stats = True
        c.Limit = 3000
        twint.run.Search(c)
        df = twint.storage.panda.Tweets_df
    return df 
df = tweet_twint(keyword_list)        
    
```

```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Inflation'])
# Interest by Region
df = pytrend.interest_by_region()
df.head(10)
```

```python
df.columns
```

```python
df.sort_values(by=['Inflation'], ascending=False)
```

```python
df.reset_index().plot(x='geoName', y='Inflation', figsize=(120, 10), kind ='bar')
```

```python
df = pytrend.trending_searches(pn='united_states')
df.head()
```

```python

#Let us see what was trending in 2019. With the help of top_charts method we can get the top trending searches yearly.

df = pytrend.top_charts(2019, hl='en-US', tz=300, geo='GLOBAL')
df.head()
```

# Let's explore on related queries

```python
pytrend.build_payload(kw_list=['Inflation'],timeframe='2020-01-01 2020-02-01')#timeframe = '2020-01-01 2020-12-31')    'today 5-y'
```

```python
# Related Queries, returns a dictionary of dataframes
related_queries = pytrend.related_queries()
```

```python
import pprint

#pprint.pprint(related_queries)
display(related_queries['Inflation']['top'].head())
related_queries['Inflation']['rising'].head()
```

```python
# Related Topics, returns a dictionary of dataframes
related_topic = pytrend.related_topics()
related_topic.values()
display(related_topic['Inflation']['top'].head())
related_topic['Inflation']['rising'].head()
```
